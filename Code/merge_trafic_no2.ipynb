# DATASET OF NO2

```python
dfn= pd.read_csv("/home/fitec/Bureau/bike_project/Bike_project/No2_data_parStaion/DonneeNo2_Paris_Station.csv",delimiter=",", encoding = "utf-8")
dfn.tail()
```

```python
dfn_null=dfn[dfn["taux_no2"].isnull()]
dfn_null
```

## REPLACE NAN VALUE BY 0

```python
dfn['taux_no2'] = dfn['taux_no2'].fillna(0)
```

## NO2 TO NUMERIC

```python
dfn["taux_no2"] = pd.to_numeric(dfn["taux_no2"], errors='coerce')
```

```python
print(dfn["date"].max())
print(dfn["date"].min())
```

## SELECT DATE AFTER 11/20

```python
dfn=dfn[dfn["date"]>="2019-11-01"]
```

## ADD WEEKDAY

```python
# pd.to_datetime(dfn['date'])
dfn['date'] = pd.to_datetime(dfn['date'], errors='coerce')
```

```python
dfn['day_of_week'] = dfn['date'].dt.day_name()
```

### group by no2 day & zip code  ( moyenne de polution par heure chauqe jour par arrondissement )

```python
dfn_gp_day= dfn.groupby(['zip_code',"date","day_of_week"]) \
       .agg({'taux_no2':'mean'}).reset_index()
dfn_gp_day.head(20)
```

```python
dfn_gp_day=dfn_gp_day.rename(columns={'date': 'date_full'})
dfn_gp_day.head()
```

```python
dfn_gp_day.sort_values(by=['date_full'])
```

```python
new_df = pd.merge(dfn_gp_day, dft_gp_day, on=['date_full','zip_code'],how='inner')
```
